```python
import pandas as pd

dataset=pd.read_csv('/content/drive/MyDrive/machine learning/DAY 6 (Regression)/none-linear-dataset.csv').values
print(dataset)
```

```python
data=dataset[:,0].reshape(-1,1)
target=dataset[:,1]
print('data:',data)
print('target:',target)
```

```python
from matplotlib import pyplot as plt
plt.scatter(data,target)
plt.xlabel('Feature -X)')
plt.ylabel('Label -Y)')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly=PolynomialFeatures(degree=2,include_bias=False)
data_new=poly.fit_transform(data)  # X variable eka digree 2 (x warga) wlt transform wela fit wenw

print('data_new:',data_new)
```

```python
from sklearn.model_selection import train_test_split
train_data,test_data,train_target,test_target=train_test_split(data_new,target,test_size=0.1)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(train_data,train_target)
```

```python
print('coefs:',model.coef_)
print('c:',model.intercept_)
```
